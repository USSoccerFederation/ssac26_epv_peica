```python
from kloppy import pff

# 10517 is the world cup final.
tracking_datasets = pff.load_tracking(
    meta_data="meta_10517.json",
    raw_data="tracking_10517.jsonl.bz2",
    roster_meta_data="roster_10517.json",
)
```

```python
for segment in TRACKING_SEGMENTS:
    print(
        tracking_datasets.filter(
            lambda frame: frame.timestamp >= segment["timestamp"]
            and frame.timestamp <= segment["end_timestamp"]
            and frame.period.id == segment["period_id"]
        )
    )
```
